```python
import pandas as pd
import numpy as np
```

```python
df_customers = pd.read_csv('../data/customers.csv')
df_payments = pd.read_csv('../data/payments.csv', sep='^')
df_orders = pd.read_json('../data/orders.json')
df_products = pd.read_excel('../data/products.xlsx', engine='openpyxl')
df_events = pd.read_xml('../data/events.xml')

print("ТАБЛИЦА КЛИЕНТОВ")
display(df_customers.head(10))

print("\nТАБЛИЦА ПЛАТЕЖЕЙ")
display(df_payments.head(10))

print("\nТАБЛИЦА ЗАКАЗОВ")
display(df_orders.head(10))

print("\nТАБЛИЦА ТОВАРОВ")
display(df_products.head(10))

print("\nТАБЛИЦА СОБЫТИЙ")
display(df_events.head(10))
```

Задание 1, избавлюсь от дубликатов и приведу все пропуски, nan, unknown к единому формату

```python
df_customers_clean = df_customers.copy()
df_payments_clean = df_payments.copy()
df_orders_clean = df_orders.copy()
df_products_clean = df_products.copy()
df_events_clean = df_events.copy()

dict_of_dfs = {
    'customers': df_customers_clean,
    'payments': df_payments_clean,
    'orders': df_orders_clean,
    'products': df_products_clean,
    'events': df_events_clean
}

primary_keys = {
    'customers': 'customer_id',
    'products': 'product_id',
    'orders': 'order_id',
    'payments': 'payment_id',
    'events': 'event_id'
}

for name, df in dict_of_dfs.items():
    df = df.drop_duplicates().copy()
    pk = primary_keys.get(name)

    if pk in df.columns:

        dup_count = df[pk].duplicated().sum()

        if dup_count > 0:
            print(f'{name}: найдено дублей по {pk}: {dup_count}')

            df = df.drop_duplicates(
                subset=[pk],
                keep='first'
            )
        df = df[df[pk].notna()].copy()
    for column in df.columns:

        df[column] = df[column].replace(
            [
                'UNKNOWN',
                'unknown',
                'UNKNOWN ',
                'unknown ',
                'nan',
                'NaN',
                ''
            ],
            pd.NA
        )

        if df[column].dtype == 'object':

            df[column] = (
                df[column]
                .astype('string')
                .str.strip()
            )

        else:

            if column.endswith('_id'):
                continue
            else:
                df[column] = df[column].fillna(0)

    dict_of_dfs[name] = df

print("Success")
```

Задание 2. Учту возможные некорретные даты, нечисловые выражения в полях с числами

```python
# некорректные даты
dq_errors = []
date_columns = ['created_at', 'payment_timestamp', 'order_timestamp', 'event_timestamp']
numeric_columns = ['amount', 'price', 'quantity', 'sum']

for table_name, df in dict_of_dfs.items():
    for col in df.columns:
        if col in date_columns:
            # Пробуем конвертировать в дату. Ошибки (2025-99-99) превратятся в NaT
            converted_date = pd.to_datetime(df[col], errors='coerce')
            bad_rows = df[converted_date.isna() & df[col].notna()]

            for idx, row in bad_rows.iterrows():
                dq_errors.append({
                    'table_name': table_name,
                    'column_name': col,
                    'error_type': 'Invalid date format',
                    'bad_value': str(row[col])
                })
            
            # Оставляем в таблице только валидные даты (без NaT)
            df[col] = converted_date
            dict_of_dfs[table_name] = df[df[col].notna()].copy()

# ищем 'error_amount'
for table_name, df in dict_of_dfs.items():
    for col in df.columns:
        if any(x in col.lower() for x in numeric_columns) and df[col].dtype == 'object':
            converted_numeric = pd.to_numeric(df[col], errors='coerce')
            bad_rows = df[converted_numeric.isna() & df[col].notna() & (df[col] != 'Unknown')]
            
            for idx, row in bad_rows.iterrows():
                dq_errors.append({
                    'table_name': table_name,
                    'column_name': col,
                    'error_type': 'Non-numeric value in financial column',
                    'bad_value': str(row[col])
                })
            
            # Оставляем только валидные числа
            df[col] = converted_numeric
            dict_of_dfs[table_name] = df[df[col].notna()].copy()

df_dq_log = pd.DataFrame(dq_errors)
df_dq_log.to_csv('../data/data_errors.csv', index=False)

print(f"-> Обнаружено критических ошибок: {len(df_dq_log)} шт.")
```
