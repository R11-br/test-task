```python
import pandas as pd
data = {
    'id': [101, 102, 103, 104],
    'name': [
        'Aндрей', # первая буква A на англ.
        'Иpa', #вторая и третья буква на англ.
        'Игорь', # на русском
        'Игoрь' # o на англ.
    ]
}
df = pd.DataFrame(data)
display(df)
```

```python
letters = {
    # английская -> русская)
    'a': 'а', 'c': 'с', 'e': 'е', 'o': 'о', 'p': 'р', 'x': 'х', 'y': 'у',
    
    # английская -> русская)
    'A': 'А', 'B': 'В', 'C': 'С', 'E': 'Е', 'H': 'Н', 'K': 'К', 
    'M': 'М', 'O': 'О', 'P': 'Р', 'T': 'Т', 'X': 'Х', 'Y': 'У'
}
trans_table = str.maketrans(letters)
df['update_name'] = df['name'].str.translate(trans_table)
df = df[['id', 'update_name']]
display(df)
```

```python
df['new_id'] = df.groupby('update_name').ngroup()
display(df)
```
